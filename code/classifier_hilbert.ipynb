```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from torchvision.utils import save_image
from torch.utils.data import DataLoader
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt

# Set random seed for reproducibility
torch.manual_seed(42)

# Define device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define directories
real_data_dirs = {
    'non_dysarthric': r"D:/datasets/Hilbert_spectrum/non_dysarthric",
    'dysarthric': r"D:/datasets/Hilbert_spectrum/dysarthric"
}

synthetic_data_dirs = {
    'non_dysarthric': r"D:/datasets/Hilbert_spectrum/synthetic/non_dysarthric",
    'dysarthric': r"D:/datasets/Hilbert_spectrum/synthetic/dysarthric"
}

# Create synthetic directories if they don't exist
for dir_path in synthetic_data_dirs.values():
    os.makedirs(dir_path, exist_ok=True)

# Image preprocessing transformations
transform = transforms.Compose([
    transforms.Resize((32, 32)),  # Further reduced image size for efficiency
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)  # Normalize to [-1, 1]
])

# Custom Dataset to load grayscale images and convert to RGB
class HilbertSpectrumDataset(torch.utils.data.Dataset):
    def __init__(self, root_dir, transform=None):
        super(HilbertSpectrumDataset, self).__init__()
        self.root_dir = root_dir
        self.transform = transform
        self.image_paths = []
        for filename in os.listdir(root_dir):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                self.image_paths.append(os.path.join(root_dir, filename))

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        try:
            image = Image.open(img_path).convert('RGB')  # Convert to RGB
            if self.transform:
                image = self.transform(image)
            return image
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            # Return a black image in case of error
            return torch.zeros(3, 32, 32)

# Lightweight DCGAN Generator
class Generator(nn.Module):
    def __init__(self, nz=100, ngf=16, nc=3):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            # Input Z
            nn.ConvTranspose2d(nz, ngf * 4, 4, 1, 0, bias=False),  # (ngf*4) x 4 x 4
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            # State size: (ngf*4) x 4 x 4
            nn.ConvTranspose2d(ngf * 4, ngf * 2, 4, 2, 1, bias=False),  # (ngf*2) x 8 x 8
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            # State size: (ngf*2) x 8 x 8
            nn.ConvTranspose2d(ngf * 2, ngf, 4, 2, 1, bias=False),      # (ngf) x 16 x 16
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            # State size: (ngf) x 16 x 16
            nn.ConvTranspose2d(ngf, nc, 4, 2, 1, bias=False),          # (nc) x 32 x 32
            nn.Tanh()
            # Output size: (nc) x 32 x 32
        )

    def forward(self, input):
        return self.main(input)

# Lightweight DCGAN Discriminator
class Discriminator(nn.Module):
    def __init__(self, nc=3, ndf=16):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            # Input size: (nc) x 32 x 32
            nn.Conv2d(nc, ndf, 4, 2, 1, bias=False),   # (ndf) x 16 x 16
            nn.LeakyReLU(0.2, inplace=True),
            # State size: (ndf) x 16 x 16
            nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias=False),  # (ndf*2) x 8 x 8
            nn.BatchNorm2d(ndf * 2),
            nn.LeakyReLU(0.2, inplace=True),
            # State size: (ndf*2) x 8 x 8
            nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias=False),  # (ndf*4) x 4 x 4
            nn.BatchNorm2d(ndf * 4),
            nn.LeakyReLU(0.2, inplace=True),
            # State size: (ndf*4) x 4 x 4
            nn.Conv2d(ndf * 4, 1, 4, 1, 0, bias=False),   # (1) x 1 x 1
            nn.Sigmoid()
            # Output: Scalar probability
        )

    def forward(self, input):
        output = self.main(input)
        return output.view(-1)  # Flatten to (batch_size)

# Function to train DCGAN for a specific class
def train_dcgan(class_name, real_dir, synthetic_dir, num_epochs=30, batch_size=32, lr=0.0002, beta1=0.5, nz=100, ngf=16, ndf=16):
    print(f"\nTraining DCGAN for class: {class_name}")
    
    # Create dataset and dataloader
    dataset = HilbertSpectrumDataset(real_dir, transform=transform)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=0, pin_memory=False)
    
    # Dataset statistics before augmentation
    real_count = len(dataset)
    print(f"Number of real images for class '{class_name}': {real_count}")
    
    # Initialize generator and discriminator
    netG = Generator(nz=nz, ngf=ngf, nc=3).to(device)
    netD = Discriminator(nc=3, ndf=ndf).to(device)
    
    # Initialize weights
    def weights_init(m):
        classname = m.__class__.__name__
        if classname.find('Conv') != -1:
            nn.init.normal_(m.weight.data, 0.0, 0.02)
        elif classname.find('BatchNorm') != -1:
            nn.init.normal_(m.weight.data, 1.0, 0.02)
            nn.init.constant_(m.bias.data, 0)
    
    netG.apply(weights_init)
    netD.apply(weights_init)
    
    # Loss function
    criterion = nn.BCELoss()
    
    # Setup Adam optimizers
    optimizerD = optim.Adam(netD.parameters(), lr=lr, betas=(beta1, 0.999))
    optimizerG = optim.Adam(netG.parameters(), lr=lr, betas=(beta1, 0.999))
    
    # Labels
    real_label = 1.
    fake_label = 0.
    
    # Training Loop
    for epoch in range(num_epochs):
        for i, data in enumerate(tqdm(dataloader, desc=f"Epoch [{epoch+1}/{num_epochs}]")):
            # Update Discriminator
            netD.zero_grad()
            real_cpu = data.to(device)
            b_size = real_cpu.size(0)
            label = torch.full((b_size,), real_label, dtype=torch.float, device=device)
            
            output = netD(real_cpu)
            # Ensure output and label have the same shape
            if output.shape != label.shape:
                print(f"Shape mismatch: output shape {output.shape}, label shape {label.shape}")
                continue
            errD_real = criterion(output, label)
            errD_real.backward()
            D_x = output.mean().item()
            
            noise = torch.randn(b_size, nz, 1, 1, device=device)
            fake = netG(noise)
            label.fill_(fake_label)
            output = netD(fake.detach())
            if output.shape != label.shape:
                print(f"Shape mismatch: output shape {output.shape}, label shape {label.shape}")
                continue
            errD_fake = criterion(output, label)
            errD_fake.backward()
            D_G_z1 = output.mean().item()
            
            errD = errD_real + errD_fake
            optimizerD.step()
            
            # Update Generator
            netG.zero_grad()
            label.fill_(real_label)  # Generator wants discriminator to believe fake is real
            output = netD(fake)
            if output.shape != label.shape:
                print(f"Shape mismatch: output shape {output.shape}, label shape {label.shape}")
                continue
            errG = criterion(output, label)
            errG.backward()
            D_G_z2 = output.mean().item()
            optimizerG.step()
        
        print(f"[{epoch+1}/{num_epochs}] Loss_D: {errD.item():.4f} Loss_G: {errG.item():.4f} "
              f"D(x): {D_x:.4f} D(G(z)): {D_G_z1:.4f}/{D_G_z2:.4f}")
    
    # Generate synthetic images after training
    netG.eval()
    num_synthetic = 500  # Number of synthetic images to generate
    with torch.no_grad():
        for i in range(num_synthetic):
            noise = torch.randn(1, nz, 1, 1, device=device)
            fake_image = netG(noise).detach().cpu()
            fake_image = (fake_image + 1) / 2  # Denormalize to [0,1]
            save_image(fake_image, os.path.join(synthetic_dir, f"{class_name}_synthetic_{i+1}.png"))
            if (i+1) % 100 == 0:
                print(f"Generated {i+1}/{num_synthetic} synthetic images for class: {class_name}")
    
    # Dataset statistics after augmentation
    synthetic_count = len(os.listdir(synthetic_dir))
    print(f"Number of synthetic images for class '{class_name}': {synthetic_count}\n")
    print(f"Completed training and synthetic data generation for class: {class_name}\n")

# Visualization of real and synthetic images (Optional)
def visualize_real_synthetic(class_name, real_dir, synthetic_dir, num_samples=5):
    real_images = [f for f in os.listdir(real_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))][:num_samples]
    synthetic_images = [f for f in os.listdir(synthetic_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))][:num_samples]
    
    fig, axes = plt.subplots(2, num_samples, figsize=(15, 6))
    for i in range(num_samples):
        # Real images
        img_path = os.path.join(real_dir, real_images[i])
        try:
            image = Image.open(img_path).convert('RGB')
            axes[0, i].imshow(image)
            axes[0, i].axis('off')
            if i == 0:
                axes[0, i].set_title(f"Real {class_name}")
        except Exception as e:
            print(f"Error loading real image {img_path}: {e}")
            axes[0, i].axis('off')
        
        # Synthetic images
        synth_path = os.path.join(synthetic_dir, synthetic_images[i])
        try:
            synth_image = Image.open(synth_path).convert('RGB')
            axes[1, i].imshow(synth_image)
            axes[1, i].axis('off')
            if i == 0:
                axes[1, i].set_title(f"Synthetic {class_name}")
        except Exception as e:
            print(f"Error loading synthetic image {synth_path}: {e}")
            axes[1, i].axis('off')
    
    plt.show()

if __name__ == "__main__":
    # Train DCGAN for both classes
    for class_name, real_dir in real_data_dirs.items():
        synthetic_dir = synthetic_data_dirs[class_name]
        train_dcgan(class_name, real_dir, synthetic_dir, num_epochs=30, batch_size=32, lr=0.0002, beta1=0.5, nz=100, ngf=16, ndf=16)
    
    # Visualize samples for each class after training
    for class_name in real_data_dirs.keys():
        visualize_real_synthetic(class_name, real_data_dirs[class_name], synthetic_data_dirs[class_name], num_samples=5)
    
    print("Data preprocessing and Generative AI (Lightweight DCGAN) completed successfully.")
```

```python
import os
import shutil
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms, datasets
from torch.utils.data import DataLoader, random_split
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import timm

# ---------------------------- #
# 1. Define Paths and Parameters
# ---------------------------- #

# Base dataset directory
base_dir = r"D:/datasets/Hilbert_spectrum"

# Real image directories
real_dysarthric_dir = os.path.join(base_dir, "dysarthric")
real_nondysarthric_dir = os.path.join(base_dir, "non_dysarthric")

# Synthetic image directories
synthetic_dysarthric_dir = os.path.join(base_dir, "synthetic", "dysarthric")
synthetic_nondysarthric_dir = os.path.join(base_dir, "synthetic", "non_dysarthric")

# Combined training directory
combined_train_dir = os.path.join(base_dir, "train")
train_dysarthric_dir = os.path.join(combined_train_dir, "dysarthric")
train_nondysarthric_dir = os.path.join(combined_train_dir, "non_dysarthric")

# Validation and Test directories
val_dir = os.path.join(base_dir, "val")
val_dysarthric_dir = os.path.join(val_dir, "dysarthric")
val_nondysarthric_dir = os.path.join(val_dir, "non_dysarthric")

test_dir = os.path.join(base_dir, "test")
test_dysarthric_dir = os.path.join(test_dir, "dysarthric")
test_nondysarthric_dir = os.path.join(test_dir, "non_dysarthric")

# Create necessary directories
for directory in [train_dysarthric_dir, train_nondysarthric_dir,
                 val_dysarthric_dir, val_nondysarthric_dir,
                 test_dysarthric_dir, test_nondysarthric_dir]:
    os.makedirs(directory, exist_ok=True)

# Parameters
image_size = 224       # Updated image size to match model expectations
batch_size = 32        # Adjust based on CPU capability
num_epochs = 30
learning_rate = 1e-4
step_size = 10         # For scheduler
gamma = 0.1            # For scheduler
random_seed = 42

# Set random seed for reproducibility
torch.manual_seed(random_seed)

# ---------------------------- #
# 2. Organize the Dataset
# ---------------------------- #

def split_and_copy_images(source_dir, train_dir, val_dir, test_dir, split_ratio=(0.7, 0.15, 0.15)):
    """
    Splits images from source_dir into train, val, and test directories based on split_ratio.
    Copies the images into respective directories.
    """
    images = [f for f in os.listdir(source_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
    total_images = len(images)
    train_end = int(split_ratio[0] * total_images)
    val_end = train_end + int(split_ratio[1] * total_images)
    
    train_images = images[:train_end]
    val_images = images[train_end:val_end]
    test_images = images[val_end:]
    
    # Function to copy images
    def copy_images(image_list, source, destination):
        for img in tqdm(image_list, desc=f"Copying to {destination}"):
            src_path = os.path.join(source, img)
            dst_path = os.path.join(destination, img)
            try:
                shutil.copyfile(src_path, dst_path)
            except Exception as e:
                print(f"Error copying {src_path} to {dst_path}: {e}")
    
    # Copy images
    copy_images(train_images, source_dir, train_dir)
    copy_images(val_images, source_dir, val_dir)
    copy_images(test_images, source_dir, test_dir)

# Split real dysarthric images
split_and_copy_images(real_dysarthric_dir, train_dysarthric_dir, val_dysarthric_dir, test_dysarthric_dir)

# Split real non-dysarthric images
split_and_copy_images(real_nondysarthric_dir, train_nondysarthric_dir, val_nondysarthric_dir, test_nondysarthric_dir)

# Copy synthetic images to training directories
def copy_synthetic_images(source_dir, target_dir):
    """
    Copies all synthetic images from source_dir to target_dir.
    """
    images = [f for f in os.listdir(source_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
    for img in tqdm(images, desc=f"Copying synthetic images to {target_dir}"):
        src_path = os.path.join(source_dir, img)
        dst_path = os.path.join(target_dir, img)
        try:
            shutil.copyfile(src_path, dst_path)
        except Exception as e:
            print(f"Error copying {src_path} to {dst_path}: {e}")

# Copy synthetic dysarthric images
copy_synthetic_images(synthetic_dysarthric_dir, train_dysarthric_dir)

# Copy synthetic non-dysarthric images
copy_synthetic_images(synthetic_nondysarthric_dir, train_nondysarthric_dir)

print("\nDataset organization completed successfully.")

# ---------------------------- #
# 3. Define Data Transformations
# ---------------------------- #

# Data augmentation and normalization for training
transform_train = transforms.Compose([
    transforms.Resize((image_size, image_size)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)  # Normalize to [-1, 1]
])

# Only resizing and normalization for validation and testing
transform_val_test = transforms.Compose([
    transforms.Resize((image_size, image_size)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

# ---------------------------- #
# 4. Create Datasets and DataLoaders
# ---------------------------- #

# Create datasets using ImageFolder
train_dataset = datasets.ImageFolder(root=combined_train_dir, transform=transform_train)
val_dataset = datasets.ImageFolder(root=val_dir, transform=transform_val_test)
test_dataset = datasets.ImageFolder(root=test_dir, transform=transform_val_test)

# Define DataLoaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0, pin_memory=False)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=0, pin_memory=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=0, pin_memory=False)

print(f"\nNumber of training samples: {len(train_dataset)}")
print(f"Number of validation samples: {len(val_dataset)}")
print(f"Number of testing samples: {len(test_dataset)}")

# ---------------------------- #
# 5. Define and Modify the Classification Model
# ---------------------------- #

# Load a pre-trained Vision Transformer (ViT-Tiny)
model_name = 'vit_tiny_patch16_224'  # Updated model name
try:
    model = timm.create_model(model_name, pretrained=True)
    print(f"\nLoaded model: {model_name}")
except RuntimeError as e:
    print(f"\nError loading model '{model_name}': {e}")
    print("Attempting to list available models...")
    available_models = timm.list_models()
    print("Available models:")
    for name in available_models:
        print(name)
    # Suggest an alternative model
    alternative_model_name = 'vit_small_patch16_224'  # Change as per available models
    if alternative_model_name in available_models:
        print(f"\nUsing alternative model: {alternative_model_name}")
        model = timm.create_model(alternative_model_name, pretrained=True)
    else:
        print("\nNo suitable alternative ViT model found. Exiting.")
        exit(1)

# Modify the classifier head for binary classification
num_ftrs = model.head.in_features
model.head = nn.Linear(num_ftrs, 2)  # Binary classification: dysarthric vs. non_dysarthric

# Move the model to the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

print("\nModel Summary:")
print(model)

# ---------------------------- #
# 6. Set Up Loss Function, Optimizer, and Scheduler
# ---------------------------- #

# Define Loss Function
criterion = nn.CrossEntropyLoss()

# Define Optimizer with different learning rates for different parts
optimizer = optim.AdamW([
    {'params': model.patch_embed.parameters(), 'lr': learning_rate},
    {'params': model.cls_token, 'lr': learning_rate},
    {'params': model.pos_drop.parameters(), 'lr': learning_rate},
    {'params': model.blocks.parameters(), 'lr': learning_rate},
    {'params': model.norm.parameters(), 'lr': learning_rate},
    {'params': model.head.parameters(), 'lr': learning_rate*10}  # Higher lr for the classifier head
], weight_decay=1e-2)

# Define Learning Rate Scheduler
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=step_size, gamma=gamma)

# ---------------------------- #
# 7. Implement Training and Validation Loops
# ---------------------------- #

def train_model_vit(model, criterion, optimizer, scheduler, num_epochs=30):
    best_f1 = 0.0
    best_model_wts = None
    history = {
        'train_loss': [],
        'val_loss': [],
        'train_acc': [],
        'val_acc': [],
        'train_f1': [],
        'val_f1': []
    }
    
    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")
        print("-" * 10)
        
        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Set model to training mode
                dataloader = train_loader
            else:
                model.eval()   # Set model to evaluate mode
                dataloader = val_loader
            
            running_loss = 0.0
            running_corrects = 0
            all_preds = []
            all_labels = []
            
            # Iterate over data
            for inputs, labels in tqdm(dataloader, desc=f"{phase.capitalize()}"):
                inputs = inputs.to(device)
                labels = labels.to(device)
                
                # Zero the parameter gradients
                optimizer.zero_grad()
                
                # Forward
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)
                
                    # Backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()
                
                # Statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
            
            if phase == 'train':
                scheduler.step()
            
            epoch_loss = running_loss / len(dataloader.dataset)
            epoch_acc = running_corrects.double() / len(dataloader.dataset)
            epoch_f1 = f1_score(all_labels, all_preds, average='weighted')
            
            history[f'{phase}_loss'].append(epoch_loss)
            history[f'{phase}_acc'].append(epoch_acc.item())
            history[f'{phase}_f1'].append(epoch_f1)
            
            print(f"{phase.capitalize()} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f} F1: {epoch_f1:.4f}")
            
            # Deep copy the model
            if phase == 'val' and epoch_f1 > best_f1:
                best_f1 = epoch_f1
                best_model_wts = model.state_dict()
                torch.save(best_model_wts, 'best_vit_model.pth')
                print(f"Best model updated with F1: {best_f1:.4f}")
    
    print("\nTraining complete.")
    print(f"Best Validation F1 Score: {best_f1:.4f}")
    
    # Load best model weights
    if best_model_wts:
        model.load_state_dict(best_model_wts)
    
    return model, history

# ---------------------------- #
# 8. Evaluate the Model on the Test Set
# ---------------------------- #

def evaluate_model_vit(model, test_loader, device='cpu'):
    model.eval()
    test_loss = 0.0
    test_corrects = 0
    all_preds = []
    all_labels = []
    all_probs = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Testing"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)
            
            test_loss += loss.item() * inputs.size(0)
            test_corrects += torch.sum(preds == labels.data)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            probs = torch.softmax(outputs, dim=1)[:,1]  # Probability for class '1'
            all_probs.extend(probs.cpu().numpy())
    
    total_loss = test_loss / len(test_loader.dataset)
    total_acc = test_corrects.double() / len(test_loader.dataset)
    total_f1 = f1_score(all_labels, all_preds, average='weighted')
    total_precision = precision_score(all_labels, all_preds, average='weighted')
    total_recall = recall_score(all_labels, all_preds, average='weighted')
    try:
        total_roc_auc = roc_auc_score(all_labels, all_probs)
    except ValueError:
        total_roc_auc = float('nan')  # Handle cases with only one class present
    
    print(f"\nTest Loss: {total_loss:.4f} Acc: {total_acc:.4f}")
    print(f"Precision: {total_precision:.4f} Recall: {total_recall:.4f} F1-Score: {total_f1:.4f}")
    print(f"ROC-AUC: {total_roc_auc:.4f}")
    
    return {
        'loss': total_loss,
        'accuracy': total_acc.item(),
        'precision': total_precision,
        'recall': total_recall,
        'f1_score': total_f1,
        'roc_auc': total_roc_auc
    }

# ---------------------------- #
# 9. Plot Training Metrics (Optional)
# ---------------------------- #

def plot_metrics_vit(history):
    epochs = range(1, len(history['train_loss']) + 1)
    
    plt.figure(figsize=(12, 5))
    
    # Plot Loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['train_loss'], label='Training Loss')
    plt.plot(epochs, history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss Over Epochs')
    plt.legend()
    
    # Plot Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['train_acc'], label='Training Accuracy')
    plt.plot(epochs, history['val_acc'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Accuracy Over Epochs')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# ---------------------------- #
# 10. Confusion Matrix and ROC Curve (Optional)
# ---------------------------- #

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc

def plot_confusion_matrix_vit(model, test_loader, device='cpu'):
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Confusion Matrix"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    cm = confusion_matrix(all_labels, all_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=test_loader.dataset.classes)
    disp.plot(cmap=plt.cm.Blues)
    plt.title("Confusion Matrix")
    plt.show()

def plot_roc_curve_vit(test_labels, test_probs):
    fpr, tpr, thresholds = roc_curve(test_labels, test_probs)
    roc_auc = auc(fpr, tpr)
    
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

# ---------------------------- #
# 11. Main Execution
# ---------------------------- #

if __name__ == "__main__":
    # Train the model
    trained_model, history = train_model_vit(model, criterion, optimizer, scheduler, num_epochs=num_epochs)
    
    print("\nModel training and validation completed successfully.")
    
    # Plot training metrics
    plot_metrics_vit(history)
    
    # Evaluate the model on the test set
    test_metrics = evaluate_model_vit(trained_model, test_loader, device=device)
    
    # Plot Confusion Matrix
    plot_confusion_matrix_vit(trained_model, test_loader, device=device)
    
    # Plot ROC Curve
    # Collect test probabilities and labels
    all_test_probs = []
    all_test_labels = []
    
    trained_model.eval()
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Collecting ROC Data"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = trained_model(inputs)
            probs = torch.softmax(outputs, dim=1)[:,1]  # Probability for class '1'
            
            all_test_probs.extend(probs.cpu().numpy())
            all_test_labels.extend(labels.cpu().numpy())
    
    plot_roc_curve_vit(all_test_labels, all_test_probs)
    
    # Save the final model
    torch.save(trained_model.state_dict(), 'final_vit_model.pth')
    print("\nFinal model saved as 'final_vit_model.pth'.")
    
    # To load the model later:
    # loaded_model = timm.create_model('vit_tiny_patch16_224', pretrained=False)
    # num_ftrs = loaded_model.head.in_features
    # loaded_model.head = nn.Linear(num_ftrs, 2)
    # loaded_model.load_state_dict(torch.load('final_vit_model.pth'))
    # loaded_model = loaded_model.to(device)
    # loaded_model.eval()
    # print("Model loaded and ready for inference.")
```

```python
import sys
import io
import os
import shutil
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, Model, optimizers, callbacks
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Set UTF-8 encoding for output

# ======================
# Dataset Configuration
# ======================
train_data_path = "D:/datasets/Hilbert_spectrum"
dysarthric_folder = os.path.join(train_data_path, "dysarthric")
non_dysarthric_folder = os.path.join(train_data_path, "non_dysarthric")
train_dir = os.path.join(train_data_path, 'train')
val_dir = os.path.join(train_data_path, 'val')

# ======================
# Data Preparation
# ======================
def prepare_datasets():
    # Create directories
    for folder in [train_dir, val_dir]:
        for cls in ['dysarthric', 'non_dysarthric']:
            os.makedirs(os.path.join(folder, cls), exist_ok=True)

    # Split and copy files
    def split_data(source, target_train, target_val, test_size=0.2):
        files = [f for f in os.listdir(source) if f.endswith('.png')]
        train_files, val_files = train_test_split(files, test_size=test_size, random_state=42)
        for f in train_files:
            shutil.copy(os.path.join(source, f), os.path.join(target_train, f))
        for f in val_files:
            shutil.copy(os.path.join(source, f), os.path.join(target_val, f))

    # Process both classes
    split_data(dysarthric_folder, 
               os.path.join(train_dir, 'dysarthric'), 
               os.path.join(val_dir, 'dysarthric'))
    
    split_data(non_dysarthric_folder, 
               os.path.join(train_dir, 'non_dysarthric'), 
               os.path.join(val_dir, 'non_dysarthric'))

prepare_datasets()

# ======================
# Enhanced Preprocessing
# ======================
@tf.autograph.experimental.do_not_convert
# Fixed Preprocessing Functions
# Fixed CLAHE Enhancement
def clahe_enhancement(image):
    """TensorFlow-compatible CLAHE with proper shape handling"""
    def _clahe_fn(img):
        # Preserve channel dimension
        img = img.reshape(128, 128).astype(np.uint8)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
        enhanced = clahe.apply(img)
        return enhanced.reshape(128, 128, 1).astype(np.float32) / 255.0
    
    enhanced = tf.numpy_function(_clahe_fn, [image], tf.float32)
    enhanced.set_shape((128, 128, 1))  # Explicitly set output shape
    return enhanced

# Fixed Hilbert Augmentation
def hilbert_augment(image):
    """Shape-preserving augmentation"""
    # Random crop with size validation
    image = tf.image.random_crop(image, size=[112, 112, 1])
    image = tf.image.resize(image, [128, 128])  # Maintains channel dimension
    image = tf.ensure_shape(image, (128, 128, 1))
    
    # Contrast/brightness with shape protection
    image = tf.cond(
        tf.random.uniform(()) < 0.5,
        lambda: tf.image.random_brightness(image, 0.15),
        lambda: image
    )
    image = tf.cond(
        tf.random.uniform(()) < 0.5,
        lambda: tf.image.random_contrast(image, 0.9, 1.1),
        lambda: image
    )
    return image
# ==================
# Data Pipeline
# ==================
# Modified Data Generators
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    preprocessing_function=lambda x: clahe_enhancement(hilbert_augment(x)),
    validation_split=0.2
)

val_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    preprocessing_function=lambda x: clahe_enhancement(x)
)

# Data Generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    color_mode='grayscale',
    class_mode='binary',
    batch_size=32,
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(128, 128),
    color_mode='grayscale',
    class_mode='binary',
    batch_size=32,
    shuffle=False
)

# Fixed Hybrid Attention Layer
class HybridAttention(layers.Layer):
    def __init__(self, ratio=8):
        super().__init__()
        self.ratio = ratio
        self.spatial_conv = layers.Conv2D(1, (7,7), padding='same', activation='sigmoid')
        
    def build(self, input_shape):
        self.channels = input_shape[-1]
        self.channel_dense1 = layers.Dense(self.channels//self.ratio, activation='relu')
        self.channel_dense2 = layers.Dense(self.channels, activation='sigmoid')
        super().build(input_shape)

    def call(self, inputs):
        # Channel attention
        channel = layers.GlobalAveragePooling2D()(inputs)
        channel = self.channel_dense1(channel)
        channel = self.channel_dense2(channel)
        channel = layers.Reshape((1, 1, self.channels))(channel)
        
        # Spatial attention
        spatial = self.spatial_conv(inputs)
        return layers.Multiply()([inputs * channel, spatial])

# Updated Model Architecture
def create_optimal_model():
    inputs = layers.Input(shape=(128, 128, 1))
    
    x = layers.Conv2D(64, (5,5), padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    
    for filters in [128, 256, 512]:
        # First convolution
        x = layers.Conv2D(filters, (3,3), padding='same')(x)
        
        # Attention block
        x = HybridAttention(ratio=8)(x)
        
        # Downsample
        x = layers.MaxPooling2D(2)(x)
        x = layers.Dropout(0.3)(x)
    
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(512, activation='swish')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)
    
    return Model(inputs, outputs)
# Training Execution
model = create_optimal_model()
model.compile(
    optimizer=optimizers.AdamW(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    callbacks=[
        callbacks.EarlyStopping(patience=5),
        callbacks.ModelCheckpoint('best_model.h5')
    ],
    verbose=1
)

# ==================
# Model Evaluation
# ==================
# Save final model
model.save("dysarthric_hilbert_model.keras")

# Generate predictions
val_generator.reset()
y_pred = model.predict(val_generator)
y_pred = np.round(y_pred).astype(int)

# Classification report
print("\nClassification Report:")
print(classification_report(val_generator.classes, y_pred))

# Plot training history
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.savefig('training_metrics.png')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def build_classifier(input_shape=(128, 128, 3)):#224
    inp = layers.Input(shape=input_shape, name='input')
    x = layers.Conv2D(32, (3, 3), activation='relu', name='conv1')(inp)
    x = layers.MaxPooling2D(pool_size=(2, 2), name='maxpool1')(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', name='conv2')(x)
    x = layers.MaxPooling2D(pool_size=(2, 2), name='maxpool2')(x)
    x = layers.Conv2D(128, (3, 3), activation='relu', name='conv3')(x)
    x = layers.MaxPooling2D(pool_size=(2, 2), name='maxpool3')(x)
    x = layers.Flatten(name='flatten')(x)
    x = layers.Dense(128, activation='relu', name='dense1')(x)
    output = layers.Dense(1, activation='sigmoid', name='output')(x)
    model = models.Model(inputs=inp, outputs=output, name='dysarthric_classifier')
    return model

# Build the model
model = build_classifier()

# Print model summary and total parameter count
model.summary()
print("Total number of parameters:", model.count_params())
```
