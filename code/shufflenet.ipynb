```python
import os
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix
import seaborn as sns

# -----------------------------------------------------------------------------
# 1. Data Loading from Directory
# -----------------------------------------------------------------------------
# Folder structure:
#   D:/datasets/Hilbert_spectrum/dysarthric
#   D:/datasets/Hilbert_spectrum/non_dysarthric
data_dir = "D:/datasets/Spectrograms"

# Load the dataset using image_dataset_from_directory.
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    labels="inferred",
    label_mode="binary",   # Binary classification (0 or 1)
    image_size=(128, 128),
    batch_size=32,
    shuffle=False  # Set shuffle=False to keep a fixed order for KFold splitting.
)

# Convert the tf.data.Dataset to NumPy arrays (assuming the dataset fits in memory)
X_list, y_list = [], []
for images, labels in dataset:
    X_list.append(images.numpy())
    y_list.append(labels.numpy())
X = np.concatenate(X_list, axis=0)
y = np.concatenate(y_list, axis=0)
print(f"Loaded {X.shape[0]} images with shape {X.shape[1:]}")

# -----------------------------------------------------------------------------
# 2. Define the ShuffleNet Building Blocks
# -----------------------------------------------------------------------------
def channel_shuffle(x, groups):
    """
    Performs a channel shuffle operation.
    Splits channels into groups, transposes, and then flattens back.
    """
    # x shape: (batch, H, W, channels)
    height, width, channels = x.shape[1], x.shape[2], x.shape[3]
    if channels % groups != 0:
        raise ValueError("Number of channels must be divisible by groups.")
    group_channels = channels // groups
    # Reshape to (batch, H, W, groups, group_channels)
    x = tf.reshape(x, [-1, height, width, groups, group_channels])
    # Transpose: swap the group and channel dimensions
    x = tf.transpose(x, [0, 1, 2, 4, 3])
    # Flatten back to (batch, H, W, channels)
    x = tf.reshape(x, [-1, height, width, channels])
    return x

def shuffle_block(x, out_channels, stride, groups=2):
    """
    A simplified ShuffleNet block.
    
    For stride=1, the input is split along channels; one half is kept (shortcut)
    while the other half is processed through:
      - 1x1 pointwise conv → BN → ReLU
      - 3x3 depthwise conv (with stride) → BN
      - 1x1 pointwise conv → BN → ReLU
    Then the branches are concatenated and a channel shuffle is applied.
    
    For stride=2, two branches are built (without splitting the input) and concatenated.
    """
    in_channels = x.shape[-1]
    
    if stride == 1:
        if in_channels % 2 != 0:
            raise ValueError("Input channels must be even for stride=1.")
        split = in_channels // 2
        # Split input into two branches
        x1 = layers.Lambda(lambda z: z[:, :, :, :split])(x)
        x2 = layers.Lambda(lambda z: z[:, :, :, split:])(x)
        
        # Process branch x2
        x2 = layers.Conv2D(split, (1, 1), padding='same', use_bias=False)(x2)
        x2 = layers.BatchNormalization()(x2)
        x2 = layers.ReLU()(x2)
        x2 = layers.DepthwiseConv2D((3, 3), strides=stride, padding='same', use_bias=False)(x2)
        x2 = layers.BatchNormalization()(x2)
        x2 = layers.Conv2D(split, (1, 1), padding='same', use_bias=False)(x2)
        x2 = layers.BatchNormalization()(x2)
        x2 = layers.ReLU()(x2)
        
        out = layers.Concatenate()([x1, x2])
    else:  # stride == 2
        # Branch 1
        branch1 = layers.DepthwiseConv2D((3, 3), strides=stride, padding='same', use_bias=False)(x)
        branch1 = layers.BatchNormalization()(branch1)
        branch1 = layers.Conv2D(out_channels // 2, (1, 1), padding='same', use_bias=False)(branch1)
        branch1 = layers.BatchNormalization()(branch1)
        branch1 = layers.ReLU()(branch1)
        # Branch 2
        branch2 = layers.Conv2D(out_channels // 2, (1, 1), padding='same', use_bias=False)(x)
        branch2 = layers.BatchNormalization()(branch2)
        branch2 = layers.ReLU()(branch2)
        branch2 = layers.DepthwiseConv2D((3, 3), strides=stride, padding='same', use_bias=False)(branch2)
        branch2 = layers.BatchNormalization()(branch2)
        branch2 = layers.Conv2D(out_channels // 2, (1, 1), padding='same', use_bias=False)(branch2)
        branch2 = layers.BatchNormalization()(branch2)
        branch2 = layers.ReLU()(branch2)
        
        out = layers.Concatenate()([branch1, branch2])
        
    # Apply channel shuffle
    out = layers.Lambda(lambda z: channel_shuffle(z, groups))(out)
    return out

def build_shufflenet(input_shape=(128, 128, 3), num_classes=1):
    """
    Builds a ShuffleNet-inspired model.
    The network starts with an initial convolution and max pooling,
    then stacks several shuffle blocks (organized in stages), followed by
    global average pooling, dropout, and a final Dense layer with sigmoid activation.
    """
    inp = layers.Input(shape=input_shape)
    
    # Initial convolution block
    x = layers.Conv2D(24, (3, 3), strides=(2, 2), padding='same', use_bias=False)(inp)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='same')(x)
    
    # Stage 2: Use shuffle blocks with output channels 240
    x = shuffle_block(x, out_channels=240, stride=2, groups=2)
    x = shuffle_block(x, out_channels=240, stride=1, groups=2)
    x = shuffle_block(x, out_channels=240, stride=1, groups=2)
    
    # Stage 3: Increase output channels to 480
    x = shuffle_block(x, out_channels=480, stride=2, groups=2)
    x = shuffle_block(x, out_channels=480, stride=1, groups=2)
    x = shuffle_block(x, out_channels=480, stride=1, groups=2)
    x = shuffle_block(x, out_channels=480, stride=1, groups=2)
    
    # Stage 4: Increase output channels to 960
    x = shuffle_block(x, out_channels=960, stride=2, groups=2)
    x = shuffle_block(x, out_channels=960, stride=1, groups=2)
    
    # Global average pooling and classification head
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)
    # For binary classification, use one neuron with sigmoid activation
    out = layers.Dense(num_classes, activation='sigmoid')(x)
    
    model = models.Model(inputs=inp, outputs=out)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# -----------------------------------------------------------------------------
# 3. K-Fold Cross-Validation Setup and Training
# -----------------------------------------------------------------------------
n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

fold_no = 1
acc_per_fold = []
loss_per_fold = []

for train_index, val_index in kf.split(X):
    print(f'\n{"="*20} Fold {fold_no} {"="*20}')
    
    # Split the data into training and validation sets
    X_train, X_val = X[train_index], X[val_index]
    y_train, y_val = y[train_index], y[val_index]
    
    # Build a new instance of the ShuffleNet model
    model = build_shufflenet(input_shape=(128, 128, 3), num_classes=1)
    
    # -----------------------------------------------------------------------------
    # Define Callbacks: ModelCheckpoint and EarlyStopping
    # -----------------------------------------------------------------------------
    checkpoint_cb = callbacks.ModelCheckpoint(
        filepath=f'best_model_fold_shuffle_spectro{fold_no}.h5',  # Save the best model in HDF5 format
        monitor='val_accuracy',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    early_stop_cb = callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        verbose=1
    )
    
    # Train the model
    history = model.fit(
        X_train, y_train,
        epochs=20,
        batch_size=32,
        validation_data=(X_val, y_val),
        callbacks=[checkpoint_cb, early_stop_cb],
        verbose=1
    )
    
    # Evaluate the model on the validation set
    scores = model.evaluate(X_val, y_val, verbose=0)
    print(f'Score for fold {fold_no}: Loss = {scores[0]:.4f}, Accuracy = {scores[1]*100:.2f}%')
    acc_per_fold.append(scores[1] * 100)
    loss_per_fold.append(scores[0])
    
    # -----------------------------------------------------------------------------
    # Visualization: Accuracy and Loss Curves for the Current Fold
    # -----------------------------------------------------------------------------
    epochs_range = range(1, len(history.history['accuracy']) + 1)
    plt.figure(figsize=(12, 5))
    
    # Accuracy Plot
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy')
    plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'Fold {fold_no} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Loss Plot
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, history.history['loss'], label='Train Loss')
    plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss')
    plt.title(f'Fold {fold_no} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # -----------------------------------------------------------------------------
    # Visualization: Confusion Matrix for the Current Fold
    # -----------------------------------------------------------------------------
    y_pred = (model.predict(X_val) > 0.5).astype("int32")
    cm = confusion_matrix(y_val, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Fold {fold_no} Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
    
    fold_no += 1

# -----------------------------------------------------------------------------
# 4. Summary of Cross-Validation Performance
# -----------------------------------------------------------------------------
print("\n========================================")
print("Average scores for all folds:")
print(f"Average Accuracy: {np.mean(acc_per_fold):.2f}% (+- {np.std(acc_per_fold):.2f})")
print(f"Average Loss: {np.mean(loss_per_fold):.4f}")
print("========================================")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def channel_shuffle(x, groups):
    """Performs a channel shuffle operation."""
    height, width, channels = x.shape[1], x.shape[2], x.shape[3]
    if channels % groups != 0:
        raise ValueError("Number of channels must be divisible by groups.")
    group_channels = channels // groups
    x = tf.reshape(x, [-1, height, width, groups, group_channels])
    x = tf.transpose(x, [0, 1, 2, 4, 3])
    x = tf.reshape(x, [-1, height, width, channels])
    return x

def shuffle_block(x, out_channels, stride, groups=2):
    """A simplified ShuffleNet block."""
    in_channels = x.shape[-1]
    
    if stride == 1:
        if in_channels % 2 != 0:
            raise ValueError("Input channels must be even for stride=1.")
        split = in_channels // 2
        x1 = layers.Lambda(lambda z: z[:, :, :, :split])(x)
        x2 = layers.Lambda(lambda z: z[:, :, :, split:])(x)

        x2 = layers.Conv2D(split, (1, 1), padding='same', use_bias=False)(x2)
        x2 = layers.BatchNormalization()(x2)
        x2 = layers.ReLU()(x2)
        x2 = layers.DepthwiseConv2D((3, 3), strides=stride, padding='same', use_bias=False)(x2)
        x2 = layers.BatchNormalization()(x2)
        x2 = layers.Conv2D(split, (1, 1), padding='same', use_bias=False)(x2)
        x2 = layers.BatchNormalization()(x2)
        x2 = layers.ReLU()(x2)

        out = layers.Concatenate()([x1, x2])
    else:
        branch1 = layers.DepthwiseConv2D((3, 3), strides=stride, padding='same', use_bias=False)(x)
        branch1 = layers.BatchNormalization()(branch1)
        branch1 = layers.Conv2D(out_channels // 2, (1, 1), padding='same', use_bias=False)(branch1)
        branch1 = layers.BatchNormalization()(branch1)
        branch1 = layers.ReLU()(branch1)

        branch2 = layers.Conv2D(out_channels // 2, (1, 1), padding='same', use_bias=False)(x)
        branch2 = layers.BatchNormalization()(branch2)
        branch2 = layers.ReLU()(branch2)
        branch2 = layers.DepthwiseConv2D((3, 3), strides=stride, padding='same', use_bias=False)(branch2)
        branch2 = layers.BatchNormalization()(branch2)
        branch2 = layers.Conv2D(out_channels // 2, (1, 1), padding='same', use_bias=False)(branch2)
        branch2 = layers.BatchNormalization()(branch2)
        branch2 = layers.ReLU()(branch2)

        out = layers.Concatenate()([branch1, branch2])

    out = layers.Lambda(lambda z: channel_shuffle(z, groups))(out)
    return out

def build_shufflenet(input_shape=(128, 128, 3), num_classes=1):
    """Builds a ShuffleNet-inspired model."""
    inp = layers.Input(shape=input_shape)

    x = layers.Conv2D(24, (3, 3), strides=(2, 2), padding='same', use_bias=False)(inp)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='same')(x)

    x = shuffle_block(x, out_channels=240, stride=2, groups=2)
    x = shuffle_block(x, out_channels=240, stride=1, groups=2)
    x = shuffle_block(x, out_channels=240, stride=1, groups=2)

    x = shuffle_block(x, out_channels=480, stride=2, groups=2)
    x = shuffle_block(x, out_channels=480, stride=1, groups=2)
    x = shuffle_block(x, out_channels=480, stride=1, groups=2)
    x = shuffle_block(x, out_channels=480, stride=1, groups=2)

    x = shuffle_block(x, out_channels=960, stride=2, groups=2)
    x = shuffle_block(x, out_channels=960, stride=1, groups=2)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)
    out = layers.Dense(num_classes, activation='sigmoid')(x)

    model = models.Model(inputs=inp, outputs=out)

    return model

# Build the model
model = build_shufflenet()

# Print model summary
model.summary()
```
