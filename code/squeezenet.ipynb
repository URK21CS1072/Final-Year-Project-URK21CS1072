```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix
import seaborn as sns

# -----------------------------------------------------------------------------
# 1. Prepare the DataFrame with File Paths and Labels
# -----------------------------------------------------------------------------
data_dir = "D:/datasets/Spectrograms"
dysarthric_dir = os.path.join(data_dir, "dysarthric")
non_dysarthric_dir = os.path.join(data_dir, "non_dysarthric")

# List all image files from both directories
dys_files = [os.path.join(dysarthric_dir, fname) for fname in os.listdir(dysarthric_dir)
             if fname.lower().endswith(('.png', '.jpg', '.jpeg'))]
non_dys_files = [os.path.join(non_dysarthric_dir, fname) for fname in os.listdir(non_dysarthric_dir)
                 if fname.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Create DataFrame with file paths and labels (1 for dysarthric, 0 for non-dysarthric)
df_dys = pd.DataFrame({'filename': dys_files, 'label': 1})
df_non = pd.DataFrame({'filename': non_dys_files, 'label': 0})
df = pd.concat([df_dys, df_non]).reset_index(drop=True)

print(f"Total images: {len(df)}")

# -----------------------------------------------------------------------------
# 2. Define the SqueezeNet Architecture
# -----------------------------------------------------------------------------
def fire_module(x, squeeze, expand, name):
    """
    Implements a fire module as used in SqueezeNet.
    Applies a 1x1 convolution (squeeze) followed by parallel 1x1 and 3x3 
    convolutions (expand) whose outputs are concatenated.
    """
    x = layers.Conv2D(squeeze, (1, 1), padding='valid', activation='relu', name=name + '_squeeze')(x)
    expand1x1 = layers.Conv2D(expand, (1, 1), padding='valid', activation='relu', name=name + '_expand1x1')(x)
    expand3x3 = layers.Conv2D(expand, (3, 3), padding='same', activation='relu', name=name + '_expand3x3')(x)
    x = layers.concatenate([expand1x1, expand3x3], axis=-1, name=name + '_concat')
    return x

def build_squeezenet(input_shape=(128, 128, 3)):
    """
    Builds a SqueezeNet-inspired model for binary classification.
    """
    inp = layers.Input(shape=input_shape)
    
    # Initial convolution and pooling layers
    x = layers.Conv2D(64, (3, 3), strides=(2, 2), padding='same', activation='relu', name='conv1')(inp)
    x = layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='same', name='maxpool1')(x)
    
    # Fire modules
    x = fire_module(x, squeeze=16, expand=64, name='fire2')
    x = fire_module(x, squeeze=16, expand=64, name='fire3')
    x = fire_module(x, squeeze=32, expand=128, name='fire4')
    x = layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='same', name='maxpool4')(x)
    
    x = fire_module(x, squeeze=32, expand=128, name='fire5')
    x = fire_module(x, squeeze=48, expand=192, name='fire6')
    x = fire_module(x, squeeze=48, expand=192, name='fire7')
    x = fire_module(x, squeeze=64, expand=256, name='fire8')
    x = layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='same', name='maxpool8')(x)
    
    x = fire_module(x, squeeze=64, expand=256, name='fire9')
    x = layers.Dropout(0.5, name='dropout')(x)
    
    # Final convolution, global average pooling and sigmoid activation for binary classification
    x = layers.Conv2D(1, (1, 1), padding='valid', name='conv_final')(x)
    x = layers.GlobalAveragePooling2D(name='global_avg_pool')(x)
    output = layers.Activation('sigmoid', name='sigmoid')(x)
    
    model = models.Model(inputs=inp, outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# -----------------------------------------------------------------------------
# 3. K-Fold Cross-Validation Setup Using ImageDataGenerator
# -----------------------------------------------------------------------------
# Define parameters
img_size = (128, 128)
batch_size = 32
n_splits = 5

# Create an instance of ImageDataGenerator for training (with rescaling)
datagen = ImageDataGenerator(rescale=1./255)

# Prepare KFold splits on the DataFrame
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

fold_no = 1
acc_per_fold = []
loss_per_fold = []

for train_index, val_index in kf.split(df):
    print(f'\n{"="*20} Fold {fold_no} {"="*20}')
    
    # Split DataFrame into training and validation sets
    df_train = df.iloc[train_index]
    df_val = df.iloc[val_index]
    
    # Create generators using flow_from_dataframe.
    # Note: For flow_from_dataframe, the column "filename" should contain full paths.
    train_generator = datagen.flow_from_dataframe(
        dataframe=df_train,
        x_col='filename',
        y_col='label',
        target_size=img_size,
        color_mode='rgb',
        class_mode='raw',  # raw because we have binary labels in a continuous format
        batch_size=batch_size,
        shuffle=True
    )
    
    val_generator = datagen.flow_from_dataframe(
        dataframe=df_val,
        x_col='filename',
        y_col='label',
        target_size=img_size,
        color_mode='rgb',
        class_mode='raw',
        batch_size=batch_size,
        shuffle=False
    )
    
    # Build a new instance of the SqueezeNet model
    model = build_squeezenet(input_shape=(img_size[0], img_size[1], 3))
    
    # -----------------------------------------------------------------------------
    # Callbacks: ModelCheckpoint and EarlyStopping
    # -----------------------------------------------------------------------------
    checkpoint_cb = callbacks.ModelCheckpoint(
        filepath=f'best_model_fold_sqeeze_spectro{fold_no}.h5',  # Save best model in HDF5 format
        monitor='val_accuracy',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    early_stop_cb = callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        verbose=1
    )
    
    # -----------------------------------------------------------------------------
    # Train the Model
    # -----------------------------------------------------------------------------
    history = model.fit(
        train_generator,
        epochs=20,
        validation_data=val_generator,
        callbacks=[checkpoint_cb, early_stop_cb],
        verbose=1
    )
    
    # -----------------------------------------------------------------------------
    # Evaluate on the Validation Set
    # -----------------------------------------------------------------------------
    scores = model.evaluate(val_generator, verbose=0)
    print(f'Score for fold {fold_no}: Loss = {scores[0]:.4f}, Accuracy = {scores[1]*100:.2f}%')
    acc_per_fold.append(scores[1] * 100)
    loss_per_fold.append(scores[0])
    
    # -----------------------------------------------------------------------------
    # Visualization: Accuracy and Loss Curves for this Fold
    # -----------------------------------------------------------------------------
    epochs_range = range(1, len(history.history['accuracy']) + 1)
    plt.figure(figsize=(12, 5))
    
    # Accuracy Plot
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy')
    plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'Fold {fold_no} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Loss Plot
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, history.history['loss'], label='Train Loss')
    plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss')
    plt.title(f'Fold {fold_no} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # -----------------------------------------------------------------------------
    # Visualization: Confusion Matrix for this Fold
    # -----------------------------------------------------------------------------
    # Predict on the validation generator.
    # Since flow_from_dataframe generates data in batches, we first predict over the entire set.
    y_pred = (model.predict(val_generator) > 0.5).astype("int32")
    # Get the true labels from the generator.
    y_true = val_generator.labels.astype("int32")
    
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Fold {fold_no} Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
    
    fold_no += 1

# -----------------------------------------------------------------------------
# 4. Summary of Cross-Validation Performance
# -----------------------------------------------------------------------------
print("\n========================================")
print("Average scores for all folds:")
print(f"Average Accuracy: {np.mean(acc_per_fold):.2f}% (+- {np.std(acc_per_fold):.2f})")
print(f"Average Loss: {np.mean(loss_per_fold):.4f}")
print("========================================")
```
